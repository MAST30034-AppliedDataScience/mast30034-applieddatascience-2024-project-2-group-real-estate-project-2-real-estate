```python
import pandas as pd
population_sa2 = pd.read_excel('population_sa2.xlsx', sheet_name=['Table 1'], header =  None)['Table 1'].drop(range(6))
header = population_sa2.iloc[1]
population_sa2 = population_sa2[3:]
population_sa2.columns = header 
vic_population_sa2 = population_sa2[population_sa2['S/T name'] == 'Victoria']
vic_population_sa2 = vic_population_sa2.dropna(axis='columns')
vic_population_sa2 = vic_population_sa2[['S/T name','SA2 code','SA2 name']]
vic_population_sa2 = vic_population_sa2.reset_index()
vic_population_sa2 = vic_population_sa2.drop(columns=['index'])
vic_population_sa2
```

```python
vic_population_sa2 = vic_population_sa2.drop_duplicates(subset='SA2 code', keep='last')
vic_population_sa2
```

```python
from urllib.request import urlretrieve
import os, requests, zipfile
from io import BytesIO
dir_name = ''
relative_dir = ''

## POSTCODES

url = "https://www.matthewproctor.com/Content/postcodes/australian_postcodes.zip"
filename = "australian_postcodes"
output_dir = f"{dir_name}{relative_dir}{filename}"
req = requests.get(url)
with zipfile.ZipFile(BytesIO(req.content)) as aus_postcode_zip:
    aus_postcode_zip.extractall(output_dir)
```

```python
df = pd.read_csv('australian_postcodes/australian_postcodes.csv')
df = df[df['state']=='VIC']
df
```

```python
df = df[['SA2_CODE_2021', 'postcode']].dropna().drop_duplicates()
```

```python
df = df.drop_duplicates(subset='SA2_CODE_2021', keep='last')
```

```python
df
```

```python
df['SA2_CODE_2021'] = df['SA2_CODE_2021'].astype(int)
```

```python
vic_population_sa2['SA2 code'] = vic_population_sa2['SA2 code'].astype(int)
```

```python
vic_sa2_postcode = vic_population_sa2.merge(df, left_on='SA2 code', right_on='SA2_CODE_2021', how='inner')
vic_sa2_postcode
```

```python
## POSTCODE TO SA2 MAPPING 2011
url = "https://www.abs.gov.au/AUSSTATS/subscriber.nsf/log?openagent&1270055006_CG_POSTCODE_2011_SA2_2011.zip&1270.0.55.006&Data%20Cubes&70A3CE8A2E6F9A6BCA257A29001979B2&0&July%202011&27.06.2012&Latest"

output_dir = f"{dir_name}{relative_dir}"
req = requests.get(url)
with zipfile.ZipFile(BytesIO(req.content)) as mapping_postcode:
    mapping_postcode.extractall(output_dir)

print("Completed downloading and extracting postcodes to SA2 mapping data 2011")
```

```python
df = pd.read_excel('1270055006_CG_POSTCODE_2011_SA2_2011.xls', 
                   sheet_name=['Table 3'], header =  None)['Table 3'].drop(range(5)).dropna()
header = df.iloc[0]
df = df[1:]
df.columns = header 
df = df[['POSTCODE', 'SA2_MAINCODE_2011', 'SA2_NAME_2011']]
df = df.loc[:,~df.columns.duplicated()]
df
```

```python
df = df.drop_duplicates(subset='SA2_MAINCODE_2011', keep='last')
df
```

```python
df['SA2_MAINCODE_2011'] = df['SA2_MAINCODE_2011'].astype(int)
df
```

```python
vic_population_sa2.info()
```

```python
vic_sa2_postcode1 = vic_population_sa2.merge(df, left_on='SA2 code', right_on='SA2_MAINCODE_2011', how='inner')

vic_sa2_postcode1
```

```python
df1 = vic_sa2_postcode1.copy()
df2 = vic_sa2_postcode.copy()
```

```python
# Merging the two dataframes on 'SA2 code' and 'SA2 name'
merged_df = pd.merge(df1, df2, on=['SA2 code', 'SA2 name'], how='outer')

# Filling missing values by combining data from both datasets
# If one dataframe has a missing value and the other has it, we'll fill in the missing parts

merged_df['POSTCODE'] = merged_df['POSTCODE'].combine_first(merged_df['postcode'])
merged_df['SA2_MAINCODE_2011'] = merged_df['SA2_MAINCODE_2011'].combine_first(merged_df['SA2_CODE_2021'])

# Drop the redundant columns after merging
merged_df.drop(columns=['postcode', 'SA2_CODE_2021', 'SA2_MAINCODE_2011', 
                        'SA2_NAME_2011', 'S/T name_y'], inplace=True)

merged_df['S/T name_x'] = merged_df['S/T name_x'].fillna('Victoria')
merged_df['POSTCODE'] = merged_df['POSTCODE'].astype(int)

merged_df
```

```python
merged_df.to_csv('sa2postcode.csv', index=False)
```
