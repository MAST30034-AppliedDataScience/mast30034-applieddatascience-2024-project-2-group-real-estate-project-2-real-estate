```python
import pandas as pd
import ast
```

```python
future_dwelling_df = pd.read_csv('../../data/landing/population/future_sa2_dwelling_information.csv')
regions_df = pd.read_csv('../../data/raw/location/sa2_to_rental_suburb_groups.csv')

future_dwelling_df['code'] = future_dwelling_df['SA2 code']
future_dwelling_df
```

```python
regions_df_list = regions_df.copy()
regions_df_list['code'] = regions_df_list['code'].apply(ast.literal_eval)

exploded_regions = regions_df_list.explode('code')
exploded_regions['code'] = exploded_regions['code'].apply(pd.to_numeric, errors='coerce')
regions_with_stats = pd.merge(exploded_regions, future_dwelling_df, on='code')

regions_with_stats['occupancy_rate_2021'] = pd.to_numeric(regions_with_stats['occupancy_rate_2021'].str.replace('%', ''), errors='coerce') / 100 
regions_with_stats['occupancy_rate_2026'] = pd.to_numeric(regions_with_stats['occupancy_rate_2026'].str.replace('%', ''), errors='coerce') / 100 
regions_with_stats['occupancy_rate_2031'] = pd.to_numeric(regions_with_stats['occupancy_rate_2031'].str.replace('%', ''), errors='coerce') / 100 

regions_with_stats
```

```python
df = regions_with_stats.copy()

TITLES = ['occupancy_rate', 'couple_family_with_children', 'couple_family_without_children', 'one_parent_family',
              'other_family', 'lone_person', 'group_household', 'total_households', 'population', 'hhs']

SUM_TITLES = ['couple_family_with_children', 'couple_family_without_children', 'one_parent_family',
              'other_family', 'lone_person', 'group_household', 'total_households', 'population']
AVG_TITLES = ['occupancy_rate', 'hhs']

for title in TITLES:

    df[title + '_2021'] = pd.to_numeric(df[title + '_2021'].fillna('0').astype(str).str.replace(',', ''), errors='coerce')
    df[title + '_2026'] = pd.to_numeric(df[title + '_2026'].fillna('0').astype(str).str.replace(',', ''), errors='coerce')
    df[title + '_2031'] = pd.to_numeric(df[title + '_2031'].fillna('0').astype(str).str.replace(',', ''), errors='coerce')

    for i in range(1, 5):
        df[title + f'_20{21 + i}'] = df[title + '_2021'] +  (df[title + '_2026'] - df[title + '_2021']) * 0.2 * i

    df[title + f'_2027'] = df[title + '_2026'] +  (df[title + '_2031'] - df[title + '_2026']) * 0.2 * i

sum_titles = [f'{title}_{year}' for title in SUM_TITLES for year in range(2021, 2028)]
avg_titles = [f'{title}_{year}' for title in AVG_TITLES for year in range(2021, 2028)]

merged_final = pd.merge(df.groupby('suburbs')[sum_titles].sum().reset_index(), df.groupby('suburbs')[avg_titles].mean().reset_index(), on='suburbs')

merged_final.to_csv('../../data/raw/dwelling/future_dwelling.csv')
```
