```python
import pandas as pd
```

```python
df_popu = pd.read_csv("../data/curated/population_cleaned/df_combined_population.csv")
df_IER_IEO = pd.read_csv("../data/curated/affluence_cleaned/combined_IEO_IER.csv")
```

```python
df_combined = pd.merge(df_popu,
                        df_IER_IEO
                        , on='SA2 code',
                          how='left')
df_combined = df_combined.drop(columns=['SA2 name_y'])
df_combined = df_combined.rename(columns={'SA2 name_x': 'SA2 name'})
```

```python
df_combined.to_csv("../data/curated/combined_population_affluence.csv", index=False)
```
