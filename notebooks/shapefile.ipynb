```python
import pandas as pd
import geopandas as gpd

# Replace 'path_to_shapefile_folder' with the actual path to your folder
shapefile_path = "../data/shapefile/SAL_2021_AUST_GDA2020_SHP/SAL_2021_AUST_GDA2020.shp"

# Load the shapefile
gdf = gpd.read_file(shapefile_path)

# Check the first few rows to understand the data
print(gdf.head())

# Check the CRS (Coordinate Reference System)
print(gdf.crs)
```

```python
# Display the columns in the GeoDataFrame
print(gdf.columns)

# Get a summary of the data
print(gdf.describe())
```

```python
import matplotlib.pyplot as plt

# Plot the shapefile data
gdf.plot()

# Add title and labels
plt.title("ABS Suburbs and Localities")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

# Show the plot
plt.show()
```

```python
suburb_name = "Abbotsbury"
selected_suburb = gdf[gdf['Abbotsbury'] == suburb_name]

# Plot the selected suburb
selected_suburb.plot()
plt.title(f"Map of {suburb_name}")
plt.show()
```
