```python
import pandas as pd
# Load the entire Excel file
excel_file = "../data/socioeconomic/socioeconomic.xls"

# Load the specific sheet named 'Table 1'
df = pd.read_excel(excel_file, sheet_name='Table 1', skiprows=4)

# Display the first few rows of the table
print(df.head())
```

```python
# Assuming your column is named '2016 State Suburb (SSC) Code'
filtered_df = df[df['2016 State Suburb (SSC) Code'].astype(str).str.startswith('2')]
print(filtered_df)
```

```python
# Rename columns if necessary to match
filtered_df.rename(columns={'2016 State Suburb (SSC) Code': 'Code'}, inplace=True)
filtered_df = filtered_df[['Code', 'Decile ranking of the SSC within Australia']]
```

```python
print(filtered_df.columns)
print(vic_gdf.columns)
```

```python
print(filtered_df['Code'].dtype)
print(vic_gdf['Code'].dtype)
```

```python
filtered_df['Code'] = filtered_df['Code'].astype(str).str.strip()
vic_gdf['Code'] = vic_gdf['Code'].astype(str).str.strip()
```

```python
# Merge the GeoDataFrame with the filtered DataFrame
gdf_merged = vic_gdf.merge(filtered_df, on='Code', how='inner')  # You can use 'outer' if you want all rows

# Check the result
print(gdf_merged.head())
print(gdf_merged.shape)  # Check the number of rows in the merged DataFrame
```

```python
import numpy as np

gdf_merged.rename(columns={'Decile ranking of the SSC within Australia': 'DecileRanking'}, inplace=True)
gdf_merged['DecileRanking'] = gdf_merged['DecileRanking'].replace('-', np.nan)
```

```python


import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import geopandas as gpd

# Sample data
# gdf_merged = gpd.read_file('path_to_your_geodatafile.geojson') or similar

# Define a color map that includes a color for NaN values
cmap = plt.get_cmap('OrRd')  # Use 'OrRd' colormap or choose another
cmap.set_bad(color='lightgrey')  # Set the color for NaN values

# Plot the merged GeoDataFrame
ax = gdf_merged.plot(column='DecileRanking', cmap=cmap, legend=True, figsize=(20, 20))

# Customize the plot
plt.title('Decile Ranking of SSC within Australia')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Display the plot
plt.show()
```
