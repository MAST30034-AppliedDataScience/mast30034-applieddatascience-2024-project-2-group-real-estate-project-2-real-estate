```python
import pandas as pd
```

```python
df1 = pd.read_csv("../data/cruated/combined_population_affluence_postcode.csv")
```

```python
# drop the missing value in postcode column (caused by when sa2 code transfered to postcode)
indexes_to_drop = df1[df1['postcode'] == 0].index

df1 = df1.drop(indexes_to_drop)
```

```python
df1.columns
```

```python
# 1. sum all population number 
agg_dict = {f'{year}_population': 'sum' for year in range(2001, 2024)}

agg_dict['ERP_2023_change_number'] = 'sum'
agg_dict['natural_2023_increase_number'] = 'sum'
agg_dict['natural_2023_increase_number'] = 'sum'
agg_dict['internal_2023_migration_number'] = 'sum'
agg_dict['overseas_2023_migration_number'] = 'sum'
agg_dict['area_in_square_kilometers'] = 'sum'
agg_dict['estimated_2024_population'] = 'sum'

agg_dict['IEO_Score'] = 'mean'
agg_dict['IEO_State Rank'] = 'mean'
agg_dict['IEO_State Decile'] = 'mean'
agg_dict['IEO_State Percentile'] = 'mean'


agg_dict['IER_Usual Resident Population'] = 'sum'
agg_dict['IER_Score'] = 'mean'
agg_dict['IER_State Rank'] = 'mean'
agg_dict['IER_State Decile'] = 'mean'
agg_dict['IER_State Percentile'] = 'mean'


df_grouped = df1.groupby('postcode').agg(agg_dict).reset_index()
```

```python
df_grouped.head(20)
```

```python
df_grouped['ERP_2023_change_percentage(%)'] = ((df_grouped['ERP_2023_change_number'] / df_grouped['2023_population']) * 100).round(2)

df_grouped['population_density_2023'] = (df_grouped['2023_population'] / df_grouped['area_in_square_kilometers']).round(2)


df_grouped = df_grouped.round(2)
```

```python
df_grouped.to_csv("../data/cruated/final_cleaned_affu_popu.csv", index=False)
```
